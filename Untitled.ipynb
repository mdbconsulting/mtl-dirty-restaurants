```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

inspection = pd.read_csv('inspection.csv', encoding='utf-8')
inspection.columns = inspection.iloc[0]
inspection = inspection.reindex(inspection.index.drop(0))
inspection.head(3)
```

```python
inspection.shape
```

```python
inspection['montant'].dtype
```

```python
inspection['montant'] = inspection['montant'].str.replace(r'[$, ]', '').astype('float')
inspection['montant'].head(5)
```

```python
inspection['montant'].hist()
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)

ax.set_title("Nombre d'infractions par catégorie")
ax.set_xlabel('Catégories')
ax.set_ylabel("Nombre d'infractions")
inspection['categorie'].value_counts().plot(kind='barh')
plt.show()

```

```python
fig = plt.figure(figsize=(12,10))
sns.boxplot(x=inspection['montant'], y=inspection['categorie'])
plt.show()
```
